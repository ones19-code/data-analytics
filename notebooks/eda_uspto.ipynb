```python
import pandas as pd

df = pd.read_parquet("../data/processed/uspto/uspto_master.parquet")
df.head()

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

df_small = df.sample(5000, random_state=0)

vectorizer = TfidfVectorizer(max_features=1000, stop_words="english")
X_small = vectorizer.fit_transform(df_small["patent_abstract"].fillna(""))

X_small.shape


```

```python
from sklearn.cluster import KMeans

k = 5  
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_small)

df_small["cluster"] = clusters

df_small["cluster"].value_counts()
```

```python
import numpy as np

terms = vectorizer.get_feature_names_out()

def top_terms(cluster_id, n=15):
    idx = (df_small["cluster"] == cluster_id).values

    # Moyenne des vecteurs du cluster (converti en array)
    center = X_small[idx].mean(axis=0)
    center = np.asarray(center).ravel()   # <<< FIX ICI

    # Top indices
    top_idx = np.argsort(center)[::-1][:n]
    return [terms[i] for i in top_idx]

for c in sorted(df_small["cluster"].unique()):
    print(f"\n=== Top termes cluster {c} ===")
    print(top_terms(c))

```

```python
import pandas as pd
df = pd.read_parquet(r"C:\Users\Lenovo\Desktop\patent-project\data\processed\uspto\uspto_master.parquet")


df.head()

```

```python
yearly = df.groupby("year").size()

yearly
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(yearly, order=(1,1,1))
model_fit = model.fit()

print(model_fit.summary())
```

```python
forecast_years = [2025, 2026, 2027, 2028, 2029, 2030]
forecast = model_fit.forecast(steps=6)
forecast.index = forecast_years

forecast
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.plot(yearly.index, yearly.values, marker="o", label="Historique (2010–2014)")
plt.plot(forecast.index, forecast.values, marker="o", label="Prévision (2025–2030)")

plt.title("Prévision des dépôts de brevets USPTO (Forecast 2025–2030)")
plt.xlabel("Année")
plt.ylabel("Nombre de brevets")
plt.grid(True)
plt.legend()
plt.show()
```
